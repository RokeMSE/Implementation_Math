# inverse(A)
- Input: A là ma trận vuông.
- Output: Ma trận nghịch đảo của ma trận A ban đầu (nếu có). Trường hợp không có ma trận nghịch đảo sẽ hiện thông báo "Ma trận không khả nghịch".

```python
from fractions import Fraction
import time

class Matrix:
     def __init__(self, A):
          self.A = A
          self.n = len(A)
          self.m = len(A[0])
          
     def if_square(self):
          # Kiểm tra ma trận vuông
          if len(self.A) == len(self.A[0]):
               return 1
          return 0
     
     def swap_row(self, i, j):
          self.A[i], self.A[j] = self.A[j], self.A[i] 
          
     def add_row(self, i, j, k):
          for l in range(self.m):
               self.A[i][l] += k*self.A[j][l] # Cộng hàng j nhân với k vào hàng i
               
     def multiply_row(self, i, k):
          for l in range(self.m):
               self.A[i][l] *= k # Nhân hàng i với k

     def inverse(self):
          start_time = time.time()
          
          if not (self.if_square()):
               print("Ma trận không khả nghịch vì không vuông")
               print()
               return None
          
          # Tạo ma trận đơn vị
          I = [[1 if (i==j) else 0 for i in range(self.n)] for j in range(self.n)]
          unit_matrix = Matrix(I)
          
          # Tạo ma trận bổ sung
          # Format: (A|I)
          B = []
          for i in range(self.n):
               B.append(self.A[i] + unit_matrix.A[i])
          B = Matrix(B)

          # Áp Gauss-Jordan. Các bước thực hiện
          # B1: Tìm hàng có phần tử khác 0 ở cột i
          # B2: Chia hàng i cho phần tử ở cột i
          # B3: Loại bỏ các phần tử ở cột i trong các hàng khác (khác Gauss Elimination ở chỗ hàng trên và dưới cũng phải loại bỏ phần tử ở cột i)
          for i in range(self.n):
               # Tìm hàng có phần tử khác 0 ở cột i
               for j in range(i, self.n):
                    if B.A[j][i] != 0:
                         B.swap_row(i, j)
                         break
               # Chia hàng i cho phần tử ở cột i
               if B.A[i][i] != 0:
                    B.multiply_row(i, Fraction(1,B.A[i][i]))
               else:
                    print("Ma trận không khả nghịch") # Vì dòng i có phần tử ở cột i bằng 0
                    print()
                    return None
               # Loại bỏ các phần tử ở cột i trong các hàng khác
               for j in range(self.n):
                    if j != i:
                         B.add_row(j, i, -B.A[j][i])
          
          end_time = time.time()
          print("Manual: ", end_time - start_time)
          return B 
     
     def print_matrix(self):
          for i in range(self.n):
               for j in range(self.m):
                    print(self.A[i][j], end = " ")
               print()
          
def print_combine_matrix(A):
     # Vì ma trận luôn vuông nên mình sẽ chia ma trận ra làm 2 phần được 
     for i in range(A.n):
          for j in range(A.m//2): # In phần A
               print(A.A[i][j], end = " ")
          print("|", end = " ")
          for j in range(A.m//2, A.m): # In phần I
               print(A.A[i][j], end = " ")
          print()
     print()
     
```

## MỞ RỘNG
- Tìm hiểu hàm/ phương thức tương ứng của các thư viện và thực hiện nó, so sánh kết quả.

```python
import numpy as np
import sympy as sp

def solveWithLib(A):
    # numpy  
    start_time = time.time()
    print(np.linalg.inv(A))
    end_time = time.time()
    numpy_time = end_time - start_time
    print(f"NumPy: {numpy_time}s")
    print()
    
    # sympy
    start_time = time.time()
    A = sp.Matrix(A)
    #print(A.inv())
    # Nên dùng 
    print(A.inv(method = "LU"))
    end_time = time.time()
    sympy_time = end_time - start_time
    print(f"SymPy: {sympy_time}s")
    print() 
    
```

# Nhận xét chung:

## Thuật toán Gauss-Jordan:
### Ý tưởng chính:
Để tìm ma trận nghịch đảo $(A^{-1})$ của ma trận vuông (A), ta tiến hành các bước sau:
1. **Ghép ma trận**: Ghép ma trận (A) với ma trận đơn vị (I) để tạo thành một ma trận mở rộng (A | I).
2. **Biến đổi Gauss-Jordan**: Sử dụng các phép biến đổi hàng sơ cấp để biến đổi ma trận (A) thành ma trận đơn vị (I). Đồng thời, áp dụng các phép biến đổi tương tự lên phần ma trận đơn vị (I) để biến đổi nó thành $(A^{-1})$.
3. **Kết quả**: Sau khi hoàn thành các phép biến đổi, phần bên phải của ma trận mở rộng sẽ trở thành ma trận nghịch đảo $(A^{-1})$.
### Các bước thực hiện:
#### Tạo ma trận mở rộng (A | I)
- Ghép ma trận (A) với ma trận đơn vị (I) để tạo thành một ma trận mở rộng (B).
#### Áp dụng các phép biến đổi Gauss-Jordan
- **Bước 1**: Tìm hàng có phần tử khác 0 ở cột (i) và hoán đổi lên hàng thứ (i).
- **Bước 2**: Chia hàng thứ (i) cho phần tử ở cột (i) để phần tử đó trở thành 1.
- **Bước 3**: Sử dụng phép biến đổi hàng sơ cấp để làm cho tất cả các phần tử khác trong cột (i) trở thành 0.
#### Tách ma trận nghịch đảo
- Sau khi biến đổi xong, phần bên phải của ma trận mở rộng (B) chính là ma trận nghịch đảo $(A^{-1})$.

## Tự làm (Manual):

### Ý tưởng:

- Kiểm tra ma trận có phải là ma trận vuông hay không.
- Thực hiện các thao tác hàng như hoán đổi hàng, cộng hàng, và nhân hàng với một số.
- Tìm ma trận nghịch đảo của ma trận vuông bằng phương pháp Gauss-Jordan.
- In ma trận ra màn hình.

### Mô tả các hàm:

#### `__init__(self, A)`
- Hàm khởi tạo để tạo một đối tượng ma trận với mảng hai chiều A.
- `self.A` là ma trận, `self.n` là số hàng, và `self.m` là số cột của ma trận.

#### `if_square(self)`
- Kiểm tra ma trận có phải là ma trận vuông hay không (số hàng bằng số cột).
- Trả về 1 nếu là ma trận vuông, ngược lại trả về 0.

#### `swap_row(self, i, j)`
- Hoán đổi hàng thứ `i` và hàng thứ `j` của ma trận.

#### `add_row(self, i, j, k)`
- Cộng hàng thứ `j` nhân với `k` vào hàng thứ `i`.

#### `multiply_row(self, i, k)`
- Nhân tất cả các phần tử trong hàng thứ `i` với `k`.

#### `inverse(self)`
- Tính ma trận nghịch đảo của ma trận vuông sử dụng phương pháp Gauss-Jordan.
- Nếu ma trận không vuông hoặc không khả nghịch, trả về `None` và in ra thông báo lỗi.

**Các bước chính của phương pháp Gauss-Jordan:**
1. Tìm hàng có phần tử khác 0 ở cột `i` và hoán đổi lên hàng thứ `i`.
2. Chia hàng thứ `i` cho phần tử ở cột `i`.
3. Loại bỏ các phần tử ở cột `i` trong các hàng khác.

#### `print_matrix(self)`
- In ma trận ra màn hình.

#### `print_combine_matrix(self, A)`
- In ma trận kết hợp dạng `(A|I)` (được sử dụng trong quá trình tìm nghịch đảo) có thể dùng để dễ dàng kiểm tra quá trình tính toán.
